```python
xlnk = Xlnk()
in_buffer = xlnk.cma_array(shape=(2250,), dtype=np.uint32)
out_buffer = xlnk.cma_array(shape=(2250,), dtype=np.uint32)
```

```python
for w in range (copy_count):
    #print(w)
    in_buffer[w] = (int(ord(temp[w * 4])) << 24) | (int(ord(temp[w * 4 + 1])) << 16) | (int(ord(temp[w * 4 + 2])) << 8) | int(ord(temp[w * 4 + 3]))
    print(in_buffer[w])
```

```python
dma_ip.sendchannel.transfer(in_buffer)
```
